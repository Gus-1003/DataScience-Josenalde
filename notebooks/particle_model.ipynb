# Exemplo de simulação de modelo que descreve predição de posição de uma partícula em função do tempo

```python
def cm_to_inch(value):
    return value/2.54
```

```python
"""
Derivadas na prática: simulação de fenômeno descrito por equação dif.
Autor: Josenalde Oliveira


A velocidade do corpo (v) no tempo (t) é associada à taxa de variação do espaço (x)
percorrido em função do tempo, ou seja, v(t) = dx/dt
Problema: tem-se disponível a relação desta taxa de variação, mas não sabemos como a partícula se desloca no tempo
Solução: resolver a equação diferencial
"""

# intervalo de simulação (integração)

h = 0.01 # segundos

# amostra inicial - instante discreto: t = k

k = 1

tMax = 10

t = 0
# condicao inicial (cm)

x = []
x_0 = 10
x.append(x_0)

while (t <= tMax):
  #print(k)
  #taxa de variação do espaço percorrido no tempo
  dx_dt = 6*(t**2) -10*t + 3 # taxa de variação do espaço percorrido no tempo (MODELO)
  x.append(x[k-1] + h*(dx_dt))
  #atualizacoes
  k = k + 1
  t = t + h

# Plotar grafico

# vetor do tempo (eixo X)
import numpy as np
import math

n_samples = round(tMax/h)
t = np.arange(0, tMax+2*h, h)

# Função analítica que descreve o movimento da partícula
x_real = np.array([])
x_real = 2*np.power(t,3) - 5*np.power(t,2) + 3*t + x_0;

import matplotlib.pyplot as plt
plt.figure(figsize=(cm_to_inch(25), cm_to_inch(15)))

plt.plot(t,x, 'b', label='estimado')
plt.grid()
plt.plot(t,x_real, 'r', label='analítico');
plt.legend(loc="upper left")
plt.title('Movimento de uma partícula');
plt.xlabel('tempo (s)');
plt.ylabel('cm');
```
