```python
p = df2.pivot_table(values=['Happiness Score','Family'], index='Region', aggfunc=np.mean)
```

```python
p
```

```python
import statsmodels.api as sm
import numpy as np

Y = p['Happiness Score']
X = p['Family']
model = sm.OLS(Y,X)
results = model.fit()
results.params
```
