##### Alinhamento de dados interno irá gerar NaN onde não há correspondência. No caso das DataFrames, o mesmo ocorre para linha,coluna indexadas sem correspondência. No caso de DataFrames sem índices explicitamente definidos para linhas e colunas, sua soma será NaN.

```python
#arange equivale ao range do Python, mas retorna ndarray ao invés de list
#https://numpy.org/doc/stable/reference/generated/numpy.arange.html
df3 = pd.DataFrame(np.arange(9).reshape(3,3), columns=list('bcd'), index=['RN', 'CE', 'RS'])
df3
```

```python
df4 = pd.DataFrame(np.arange(12.).reshape(4,3), columns=list('bde'), index=['PI', 'RN', 'CE','MT'])
df4
```

```python
df3+df4
```

```python
df5 = pd.DataFrame({'A': [1,2]})
df6 = pd.DataFrame({'B': [3,4]})
df5-df6 # não há rótulo em comum nas colunas
```

##### Pode ser usado um valor para preencher os locais sem coorespondência, com valor padrão, usando os métodos aritméticos. Por exemplo, ao usar add

```python
df5.add(df6, fill_value=0)
```

##### Ao fazermos um reindex num DF, o fill_value também se aplica

```python
df3.reindex(df4.columns, fill_value=0)
```
