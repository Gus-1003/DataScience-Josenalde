```python
def mod_cols(col):
    #remove espaços em branco ao redor
    col = col.strip()
    #remove parenteses
    col = col.replace('(', '')
    col = col.replace(')', '')
    #passa para minusculo
    col = col.lower()
    #preenche espaço entre palavras com _
    col = col.replace(' ', '_')
    return col
    
df.columns = [mod_cols(c) for c in df.columns]
df.columns
#df.columns.tolist()
```

```python
df['screen_size'] = (df['screen_size'].str.replace('"', ''))
```

```python
df['screen_size'] = df['screen_size'].astype(float)
```

```python
df['screen_size']
```

```python
def mod_weight(w):
    w = w.strip()
    #alguns lugares tem kg, noutros kgs
    w = w.replace('kg', '')
    w = w.replace('s', '')
    w = float(w)
    return w

df['weight'] = [mod_weight(w) for w in df['weight']]
df['weight']
```
