```python
def label(element, x):
    if element > x:
        return 'High' # ou 1
    else:
        return 'Low' # ou 0

# para uma função com o valor de comparação já definido, map() funciona bem

def label2(element):
    if element > 1:
        return 1
    else:
        return 0
#df_15['economy_impact'] = df_15['economy'].map(label2)


df_15['economy_impact'] = df_15['economy'].apply(label, x = 1.1)
```

```python
df_15
```

```python
df_15.rename(columns={'Family': 'family', 'Health (Life Expectancy)': 'health', 'Freedom': 'freedom', 
                      'Trust (Government Corruption)': 'trust', 'Generosity': 'generosity'},inplace=True)
factors = ['economy', 'family', 'health', 'freedom', 'trust', 'generosity']
```

```python
factors_impact = df_15[factors].applymap(label2)
```

```python
factors_impact
```

```python
factors_impact.apply(pd.value_counts)
```

```python
def v_counts(col):
    n = col.value_counts()
    d = col.size
    return n/d

factors_impact.apply(v_counts)
```
