```python
import pandas as pd
import numpy as np
```

```python
df_15 = pd.read_csv('../datasets/wh2015.csv')
df_16 = pd.read_csv('../datasets/wh2016.csv')
df_17 = pd.read_csv('../datasets/wh2017.csv')
```

```python
df_15.head()
head_15 = df_15.head()
head_15 = head_15.iloc[:,0:5]
head_15['year'] = 2015
head_15
```

```python
head_16 = df_16.head()
head_16 = head_16.iloc[:,0:5]
head_16['year'] = 2016
head_16
```

```python
head_17 = df_17.head()
head_17 = head_17.iloc[:,0:5]
head_17['year'] = 2017
head_17
```

```python
df_15_16 = pd.concat([head_15.iloc[:,[0,3,5]], head_16.iloc[:,[0,3,5]]], axis=0, ignore_index=True)
df_15_16
```

```python
df_15_16_h = pd.concat([head_15.iloc[:,[0,3,5]], head_16.iloc[:,[0,3,5]]], axis=1)
df_15_16_h
```

```python
df_15_16_v = pd.concat([head_15.iloc[:,[0,3,4,5]], head_16.iloc[:,[0,3,5]]], axis=0, ignore_index=True)
df_15_16_v
```

```python
df_merge1 = pd.merge(left=head_15.iloc[0:3,[0,2,5]], right=head_16.iloc[0:2,[0,2,5]], on='Country')
df_merge1
```

```python
df_merge1 = pd.merge(left=head_15.iloc[0:3,[0,2,5]], right=head_16.iloc[0:3,[0,2,5]], on='Country', suffixes=('_2015','_2016'))
df_merge1
```

```python
df_merge2 = pd.merge(left=head_15.iloc[0:3,[0,2,5]], right=head_16.iloc[0:3,[0,2,5]], on='Country', how='left',suffixes=('_2015','_2016'))
df_merge2
```

```python
df_merge3 = pd.merge(left=head_15.iloc[0:3,[0,2,5]], right=head_16.iloc[0:3,[0,2,5]], 
                     how='left',
                     left_index=True, right_index=True,
                     suffixes=('_2015','_2016'))
df_merge3
```
