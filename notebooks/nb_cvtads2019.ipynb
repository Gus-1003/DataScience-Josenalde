```python
grid = "Semestre 1: TAD0001 - FUNDAMENTOS DA COMPUTACAO - 60h	Obrigatória	, TAD0006 - SISTEMAS OPERACIONAIS - 60h	Obrigatória, TAD0102 - ALGORITMOS E PROGRAMAÇÃO - 90h	Obrigatória, TAD0105 - MATEMÁTICA APLICADA I - 60h	Obrigatória, TAD0201 - RACIOCÍNIO LÓGICO - 45h	Obrigatória, TAD0202 - LEITURA E PRODUÇÃO DE TEXTOS - 60h, Semestre 2: TAD0009 - PROGRAMACAO ORIENTADA A OBJETOS - 60h	Obrigatória, TAD0012 - PROCESSO DE DESENVOLVIMENTO DE SOFTWARE - 45h	Obrigatória, TAD0013 - MATEMATICA APLICADA II - 60h	Obrigatória, TAD0016 - VERTENTES PRODUTIVAS NAS CIENCIAS AGRARIAS - 45h	Obrigatória, TAD0103 - BANCO DE DADOS - 60h	Obrigatória, TAD0111 - PROGRAMAÇÃO VISUAL E AUTORIA WEB - 60h	Obrigatória, TAD0114 - REDES DE COMPUTADORES - 60h, Semestre 3: TAD0004 - ANALISE E PROJETO ORIENTADO A OBJETOS - 45h	Obrigatória, TAD0019 - PROGRAMACAO WEB - 60h	Obrigatória, TAD0020 - ESTRUTURAS DE DADOS - 60h	Obrigatória, TAD0022 - ESTATISTICA APLICADA - 45h	Obrigatória, TAD0025 - INTELIGÊNCIA COMPUTACIONAL - 60h	Obrigatória, TAD0123 - SISTEMAS DIGITAIS - 45h	Obrigatória, TAD0203 - GESTÃO DE QUALIDADE DE SOFTWARE - 45h, Semestre 4: TAD0018 - PROCESSAMENTO DIGITAL DE IMAGENS - 60h	Obrigatória, TAD0027 - PROGRAMACAO PARA DISPOSITIVOS MOVEIS - 90h	Obrigatória, TAD0028 - PLANEJAMENTO E GERENCIA DE PROJETOS - 45h	Obrigatória, TAD0029 - ARQUITETURA DE SOFTWARE - 45h	Obrigatória, TAD0030 - MICROCONTROLADORES - 60h	Obrigatória, TAD0032 - METODOLOGIA DO TRABALHO CIENTIFICO - 30h	Obrigatória, TAD0204 - FUNDAMENTOS E TÉCNICAS EM CIÊNCIA DE DADOS - 60h, Semestre 5: TAD0015 - EMPREENDEDORISMO - 45h	Obrigatória, TAD0108 - CONTEXTO GEOGRÁFICO REGIONAL - 30h	Obrigatória, TAD0156 - PROJETOS APLICADOS I - 60h	Obrigatória, TAD0160 - SISTEMAS DISTRIBUÍDOS - 30h	Obrigatória, TAD0205 - TRABALHO DE CONCLUSÃO DE CURSO I - 30h, Semestre 6: TAD0035 - LEGISLACAO E ETICA EM TI - 45h	Obrigatória, TAD0140 - TRABALHO DE CONCLUSAO DE CURSO II - 60h	Obrigatória, TAD0157 - PROJETOS APLICADOS II - 60h"

# remover Obrigatória
grid_transformation_1 = grid.replace('Obrigatória', '') # adiciona \t dado o ctrl c ctrl v do sigaa
#remove \t
grid_transformation_2 = grid_transformation_1.replace('\t', '') 
#substitui vírgula por ponto e vírgula
grid_transformation_3 = grid_transformation_2.replace(',', ';')
#cria array usando o separador ;, sendo objeto iterável
grid_as_list = grid_transformation_3.split(';')
#print(grid_as_list) #apenas para debug

#usar expressões regulares para remover a palavra Semestre x:
import re

#remove espaços em branco ao redor de cada componente
for component in grid_as_list:
    grid_as_list[grid_as_list.index(component)] = grid_as_list[grid_as_list.index(component)].strip() 
        
pattern = '^Semestre [0-9]:'
s = 1 # semestre inicial
f = False
for component in grid_as_list:
    if 'Semestre' in component and f == False:
        grid_as_list[grid_as_list.index(component)] = re.sub(pattern, '', component).strip()
        component = re.sub(pattern, '', component).strip()
        f = True
    elif 'Semestre' in component and f == True:
        grid_as_list[grid_as_list.index(component)] = re.sub(pattern, '', component).strip()
        component = re.sub(pattern, '', component).strip()
        s += 1
        
    grid_as_list[grid_as_list.index(component)] = grid_as_list[grid_as_list.index(component)] + ' - ' + f'{s}'
    component = component.strip() + ' - ' + f'{s}'
                
print(grid_as_list) #apenas para debug

import pandas as pd

code = []
subject = []
ch = []
semester = []
dict_grid_components = {}
for component in grid_as_list:
    items = component.split('-')
    code.append(items[0].strip())
    subject.append(items[1].strip())
    ch.append(items[2].strip())
    semester.append(items[3].strip())
    
    
    
dict_grid_components['code'] = code
dict_grid_components['subject'] = subject
dict_grid_components['ch'] = ch
dict_grid_components['semester'] = semester
    
df_components = pd.DataFrame(data=dict_grid_components) 
df_components
```

```python
dict_requisites = {
    'TAD0009': ['TAD0102'],
    'TAD0013': ['TAD0105'],
    'TAD0019': ['TAD0009'],
    'TAD0020': ['TAD0102'],
    'TAD0025': ['TAD0102'],
    'TAD0018': ['TAD0102'],
    'TAD0027': ['TAD0009'],
    'TAD0029': ['TAD0009, TAD0111, TAD0103'], #OU
    'TAD0204': ['TAD0022, TAD0102'], #E
    'TAD0156': ['TAD0102,TAD0103,TAD0012'], #E
    'TAD0160': ['TAD0114'],
    'TAD0157': ['TAD0156, TAD0028'] #E
}

df_requisites = pd.DataFrame(data=dict_requisites)
df_requisites
```

```python
query = input('Componente de interesse: ')
codes = df_components.get('code').loc[df_components['subject']\
                          .str\
                          .contains(query.upper())]\
                          .to_string(index=False)
codes = codes.split('\n')

for i in codes:
    print('---------------------------------')
    r = df_requisites.get(i, default=pd.Series(data=[], dtype='float64')) #necessario criar series vazia para usar o empty no caso de nao ter requisito
    if (not (r.empty)):
        #print(f'r is a {type(r.values)} with {r.size} element and value: {r.values}')
        r_split = r[0].split(',') #divide a string com os requisitos separados por vírgula
        print(df_components['subject'].loc[df_components['code'] == i].to_string(index=False) + ': ')
        for j in r_split:
            print('- ' + df_components['subject'].loc[df_components['code'] == j.strip()].to_string(index=False))
    else:
        print(df_components['subject'].loc[df_components['code'] == i].to_string(index=False) + ': sem requisitos')
        continue        
```
